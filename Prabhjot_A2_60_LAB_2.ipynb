# Continued on a different dataset

```python
df=pd.read_csv("city_day.csv")
```

```python
df
```

```python
df.isnull().sum()
```

Imputing with mean

```python
df['NO'].fillna(df['NO'].mean(), inplace=True)
df['NO'].isnull().sum()
```

Imputing with median

```python
print(df['NO2'].median())
df['NO2'].fillna(df['NO2'].median(), inplace=True)
print(df['NO2'].isnull().sum())
```

Imputing with mode

```python
print(df['NO2'].mode()[0])
df['NOx'].fillna(df['NOx'].mode()[0], inplace=True)
print(df['NOx'].isnull().sum())
```
